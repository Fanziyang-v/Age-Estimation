```python
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mat = scipy.io.loadmat("datasets/IMDB-WIKI/imdb_crop/imdb.mat")
```

```python
instances = mat["imdb"][0][0][0].shape[1]
```

```python
columns = [
    "dob",
    "photo_taken",
    "full_path",
    "gender",
    "name",
    "face_location",
    "face_score",
    "second_face_score",
]
```

```python
df = pd.DataFrame(index=range(instances), columns=columns)
data = mat["imdb"][0][0]
for i, col in enumerate(columns):
    df[col] = data[i][0]
```

```python
df
```

```python
# Remove samples without face
df = df[df["face_score"] != -np.inf]
```

```python
df
```

```python
# Remove samples with more than one face
df = df[df["second_face_score"].isna()]
```

```python
df
```

```python
from datetime import datetime, timedelta

def datenum_to_datetime(datenum):
    """Convert MATLAB datenum to Python datetime."""
    date = datetime(1, 1, 1) + timedelta(days=datenum - 1)
    return date.year
```

```python
df["date_of_birth"] = df["dob"].apply(datenum_to_datetime)
```

```python
df
```

```python
df["age"] = df["photo_taken"] - df["date_of_birth"]
```

```python
df
```

```python
# Remove unnecessary columns
df = df.drop(columns=["name", "photo_taken", "face_location", "face_score", "second_face_score", "date_of_birth"])
```

```python
df
```

```python
df["age"].min(), df["age"].max()
```

```python
# Remove samples with age outside the range [0, 100]
df = df[(df["age"] >= 0) & (df["age"] <= 100)]
df
```

```python
histogram_age = df["age"].hist(bins=df["age"].nunique(), figsize=(10, 5))
```

```python
histogram_gender = df["gender"].hist(bins=df["gender"].nunique(), figsize=(10, 5))
```
